```python
# GERENCIADOR DE FUNCIONÁRIOS
'''
INTEGRANTES:
- Renato Caldara
- Keity de Moura Pinho
- Fabrício John Sullivan
- Raquel de Souza
- Lisiane Ferreira Moretto
- Caroline Lazaro
'''

lista_funcionarios = []

print('\nBem-vindos(as) ao Gerenciador de Funcionários do Grupo 3!\n')


# Função para CADASTRAR UM FUNCIONÁRIO

def cadastrar_funcionario():
    cpf = int(input("\nDigite o CPF do funcionário: "))
    nome = input("Digite o nome do funcionário: ")
    cidade = input("Digite o cidade do funcionário: ")
    estado = input("Digite o estado do funcionário: ")
    idade = int(input("Digite a idade do funcionário: "))
    escolaridade = input("Digite a escolaridade do funcionário: ")
    cargo = input("Digite o cargo do funcionário: ")
    email = input("Digite o e-mail do funcionário: ")
    salario = float(input("Digite o salário do funcionário: "))

    funcionario = {
        "cpf": cpf,
        "nome": nome,
        "cidade": cidade,
        "estado": estado,
        "idade": idade,
        "escolaridade": escolaridade,
        "cargo": cargo,
        "email": email,
        "salario": salario
    }

    lista_funcionarios.append(funcionario)


# Função para CONSULTAR DADOS de um funcionário

def consultar_funcionario():
  while True:
    print("\nOPÇÕES DE CONSULTA:")
    print("1. Consultar Todos")
    print("2. Consultar por CPF")
    print("3. Consultar por nome")
    print("4. Retornar ao menu")

    opcao = input("\nEscolha uma opção: ") # Usuário escolhe uma opção

    if opcao == "1":
      print("\nLista de todos os funcionários: \n")
      for funcionario in lista_funcionarios:
        print(funcionario)

    elif opcao == "2":
      cpf_busca = int(input("\nDigite o CPF do funcionário: "))
      encontrado = False
      for funcionario in lista_funcionarios:
        if funcionario["cpf"] == cpf_busca:
          print("\nFuncionário encontrado:")
          print(funcionario)
          encontrado = True
          break

      if not encontrado:
        print("\nFuncionário não encontrado.")

    elif opcao == "3":
      nome_busca = input("\nDigite o nome a ser consultado: ")
      print("\nNome do funcionário{}:".format(nome_busca))
      encontrado = False
      for funcionario in lista_funcionarios:
        if funcionario["nome"] == nome_busca:
         print(funcionario)
         encontrado = True
      if not encontrado:
        print("\nFuncionário não encontrado.")

    elif opcao == "4":
      break
    else:
     print("Opção inválida.")


# Função para ALTERAR DADOS de um funcionário

def alterar_funcionario():
        cpf_alterar = int(input("\nDigite o CPF a ser alterado: "))
        encontrado = False
        for funcionario in lista_funcionarios:
          if funcionario["cpf"] == cpf_alterar:
            print("\nFuncionário encontrado:")
            print(funcionario)
            encontrado = True
            index_funcionario = next(index for index, funcionario in enumerate(lista_funcionarios) if funcionario["cpf"] == cpf_alterar)
            funcionario = lista_funcionarios[index_funcionario]
            novo_cpf = int(input("\nAtualize o novo CPF (ou pressione Enter para manter o mesmo): "))
            novo_nome = input("Atualize o novo nome (ou pressione Enter para manter o mesmo): ")
            novo_cidade = input("Atualize a nova cidade (ou pressione Enter para manter o mesmo): ")
            novo_estado = input("Atualize o novo estado (ou pressione Enter para manter o mesmo): ")
            novo_idade = input("Atualize a nova idade (ou pressione Enter para manter o mesmo): ")
            novo_escolaridade = input("Atualize a nova escolaridade (ou pressione Enter para manter o mesmo): ")
            novo_cargo = input("Atualize o novo cargo (ou pressione Enter para manter o mesmo): ")
            novo_email = input("Atualize o novo email (ou pressione Enter para manter o mesmo): ")
            novo_salario = input("Atualize o novo salário (ou pressione Enter para manter o mesmo): ")

            if novo_cpf:
              funcionario["cpf"] = novo_cpf
            if novo_nome:
              funcionario["nome"] = novo_nome
            if novo_cidade:
              funcionario["cidade"] = novo_cidade
            if novo_estado:
              funcionario["estado"] = (novo_estado)
            if novo_idade:
              funcionario["idade"] = (novo_idade)
            if novo_escolaridade:
              funcionario["escolaridade"] = (novo_escolaridade)
            if novo_cargo:
              funcionario["cargo"] = novo_cargo
            if novo_email:
              funcionario["email"] = novo_email
            if novo_salario:
              funcionario["salario"] = novo_salario
            break

          if not encontrado:
              print("\nFuncionário não encontrado.")


# Função para REMOVER UM FUNCIONÁRIO

def remover_funcionario():
        cpf_remover = int(input("\nDigite o CPF do funcionario a ser removido: "))
        encontrado = False
        for funcionario in lista_funcionarios:
            if funcionario["cpf"] == cpf_remover:
                  print("\nFuncionário Excluido:")
                  print(funcionario)
                  encontrado = True
                  index_funcionario = next(index for index, funcionario in enumerate(lista_funcionarios) if funcionario["cpf"] == cpf_remover)
                  del lista_funcionarios[index_funcionario]
            if not encontrado:
              print("\nFuncionário não encontrado.")


# MENU INICIAL PARA O USUÁRIO

while True:
  print('\nESCOLHA UMA OPÇÃO:')
  print('1. Cadastrar novo um funcionário')
  print('2. Consultar informações de funcionários')
  print('3. Alterar dados de um funcionário')
  print('4. Remover um funcionário')
  print('5. Encerrar sistema')
  opcao = int(input()) # Usuário escolhe uma opção

  if opcao == 1:
    cadastrar_funcionario()
  elif opcao == 2:
    consultar_funcionario()
  elif opcao == 3:
    alterar_funcionario()
  elif opcao == 4:
    remover_funcionario()
  elif opcao == 5:
    print("\nSistema encerrado!")
    break
  else:
    print("\nOpção inválida!")
```
